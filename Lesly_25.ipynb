```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib notebook
```

```python
# Employment by Industry files
employment_women_path = "Lesly_Resources/Employment_Level_Women.csv"
employment_men_path = "Lesly_Resources/Employment_Level_Men.csv"
federal_path = "Lesly_Resources/FederalMinimumWageIncreases_1968-2019.csv"
# Read csv files
employment_women = pd.read_csv(employment_women_path)
employment_men = pd.read_csv(employment_men_path)
federal_wages = pd.read_csv(federal_path)
```

```python
# Creating DataFrame
female_df = pd.DataFrame(employment_women)
female_df.head()
```

```python
# Adding a "Gender" Column 
female_df["Gender"] = "Female"
female_df.head()
```

```python
male_df = pd.DataFrame(employment_men)
male_df.head()
```

```python
# Adding a "Gender" Column 
male_df["Gender"] = "Male"
male_df.head()
```

```python
combined_gender = pd.concat([female_df, male_df])
combined_gender.head()
```

```python
combined_gender.count()
```

```python
clean_df = combined_gender.dropna(how="any")
clean_df.count()
clean_df.head()
```

## Data Visuals

```python
df_gender = pd.DataFrame(pd.pivot_table(clean_df,columns=['Gender','Year'],aggfunc = np.mean))
df_plot = df_gender.groupby(['Year','Gender']).mean().reset_index()
sns.lineplot(data = df_plot,x = 'Year',y = 0,hue = 'Gender')
plt.title('Employment Level per Gender')
plt.ylabel('Employment Level');
```

```python
emp_df = clean_df.copy()
emp_df.head()
```

```python
year_df = emp_df.groupby(['Year']).sum()
year_df['Annual'] = year_df.sum(axis=1)
year_df=year_df.reset_index()
year_df.head()
```

```python
year_df = year_df[['Year', 'Annual']]
year_df.head()
```

```python
# Combine Year df with Federal Minimum Wage
combined_data = pd.merge(year_df, federal_wages, on="Year")
combined_data.head()
```

```python
min_wage_df = combined_data.copy()
```

```python
min_wage_df.plot("Year", "Annual", color='b', title='Labor Force Statistics')
plt.ylabel('Employment Level  (per thousands)');
```

```python
min_wage_df.plot(kind='line', x='Year', y='Federal Minimum Wage', color='red');
plt.ylabel('Federal Minimum Wage')
plt.title('Labor Force Statistics');
```
