```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Employment by Industry files
employment_women_path = "Lesly_Resources/Employment_Level_Women.csv"
employment_men_path = "Lesly_Resources/Employment_Level_Men.csv"
federal_path = "Lesly_Resources/FederalMinimumWageIncreases_1968-2019.csv"
# Read csv files
employment_women = pd.read_csv(employment_women_path)
employment_men = pd.read_csv(employment_men_path)
federal_wages = pd.read_csv(federal_path)
```

```python
# Creating DataFrame
female_df = pd.DataFrame(employment_women)
female_df.head()
```

```python
# Adding a "Gender" Column 
female_df["Gender"] = "Female"
female_df.head()
```

```python
male_df = pd.DataFrame(employment_men)
male_df.head()
```

```python
# Adding a "Gender" Column 
male_df["Gender"] = "Male"
male_df.head()
```

```python
combined_gender = pd.concat([female_df, male_df])
combined_gender.head()
```

```python
combined_gender.count()
```

```python
clean_df = combined_gender.dropna(how="any")
clean_df.count()
clean_df.head()
```

## Data Visuals

```python
var = pd.DataFrame(pd.pivot_table(clean_df,columns=['Gender','Year'],aggfunc = np.mean))
test = var.groupby(['Year','Gender']).mean().reset_index()
sns.lineplot(data = test,x = 'Year',y = 0,hue = 'Gender')
plt.title('Employment Level per Gender')
plt.ylabel('Employment Level per Thousands');
```
